```python
from IPython.display import Audio
import librosa
```

```python
original_audio_filepath = '/srv/scratch/chacmod/auskidtalk_audio/1050_task1_100.wav'
```

```python
# Load the original audio file
original_audio, original_sr = librosa.load(original_audio_filepath, sr=None)
print(f'Original audio has a sampling rate of {original_sr}')
# Load the downsampled audio file
downsampled_audio, downsampled_sr = librosa.load(original_audio_filepath, sr=16000)
print(f'Downsampled audio has a sampling rate of {downsampled_sr}')
```

```python
original_audio_sound = Audio(data=original_audio, rate=original_sr)
downsampled_audio_sound = Audio(data=downsampled_audio, rate=downsampled_sr)
```
